SkillNer is an NLP module to automatically Extract skills and certifications from unstructured job postings, texts, and applicant's resumes.

Skillner uses EMSI databse (an open source skill database) as a knowldge base linker to prevent skill duplications.

spaCy is a free open-source library for Natural Language Processing in Python. It features NER, POS tagging, dependency parsing, word vectors and more.

````python
#appending all data to a single list
data = []
for i in df['RAW DATA'][:1000]:   #trying 1000 values
  data.append(i)

````

````python


import spacy
from spacy.matcher import PhraseMatcher
from skillNer.general_params import SKILL_DB
# import skill extractor
from skillNer.skill_extractor_class import SkillExtractor

# init params of skill extractor
nlp = spacy.load("en_core_web_lg")
# init skill extractor
skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)

#passing our list as a string
# extract skills from description
description = f"""{data}"""

annotations = skill_extractor.annotate(description)
````


However, we can see that this module is still not be able to capture many technical skills. Moving on, this ouput is in HTML object.

````python
x = skill_extractor.describe(annotations)
````

````python

#since this data is in html object

#we will convert in into 
from bs4 import BeautifulSoup

html_string = x.data
soup = BeautifulSoup(html_string, 'lxml')
````

````python

#finding element which is in  the red box
e = soup.findAll("span", class_ ="relative p-1 text-white rounded-md border bg-red-500")

#getting all text values
lis = []
for x in e:
  lis.append(x.text[:45]) #since the mac length of a skills is 41, we will scrape data till the text lenght  is 45


````
